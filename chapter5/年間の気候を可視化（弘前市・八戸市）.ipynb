## 5.3 年間の気候を可視化（弘前市・八戸市）
気象データは書籍を参考に取得してください。

```python
import requests

HTML_HIROSAKI = "弘前.html"
HTML_HACHINOHE = "八戸.html"

# 気象庁ホームページ
urls = ["弘前市の平年値URL", "八戸市の平年値URL"]

# データ取得と保存のループ
for url, html_file in zip(urls, [HTML_HIROSAKI, HTML_HACHINOHE]):
    response = requests.get(url, timeout=10)
    response.encoding = response.apparent_encoding

    if response.status_code == 200:
        with open(html_file, "w", encoding="utf-8") as file:
            file.write(response.text)
        print(f"Successfully {html_file}")
    else:
        print(f"Failed {html_file}: HTTP {response.status_code}")
```

```python
import pandas as pd
from bs4 import BeautifulSoup
from io import StringIO

# 弘前市
with open(HTML_HIROSAKI, "r") as file:
    soup = BeautifulSoup(file, features="lxml")
table = soup.find("table", id="tablefix1")
df_HIROSAKI_origin = pd.read_html(StringIO(str(table)))[0]

# 八戸市
with open(HTML_HACHINOHE, "r") as file:
    soup = BeautifulSoup(file, features="lxml")
table = soup.find("table", id="tablefix1")
df_HACHINOHE_origin = pd.read_html(StringIO(str(table)))[0]
```

```python
df_HIROSAKI_origin
```

```python
df_HACHINOHE_origin
```

```python
# 列の抽出
df_HIROSAKI = df_HIROSAKI_origin.filter(
    items=["要素", "降水量 (mm)", "平均気温 (℃)"]
)
# 列名変更
df_HIROSAKI = df_HIROSAKI.rename(
    columns={"要素": "月", "降水量 (mm)": "降水量(mm)"}
)
# 行の抽出
df_HIROSAKI = df_HIROSAKI[df_HIROSAKI["月"].str.contains("月")]
# 型変換
df_HIROSAKI["降水量(mm)"] = df_HIROSAKI["降水量(mm)"].astype(float)
df_HIROSAKI["平均気温 (℃)"] = df_HIROSAKI["平均気温 (℃)"].astype(float)

df_HIROSAKI
```

```python
df_HACHINOHE_origin.columns
```

```python
# 列の抽出
df_HACHINOHE = df_HACHINOHE_origin.filter(
    items=[
        ("要素", "要素", "要素"),
        ("降水量", "合計 (mm)", "合計 (mm)"),
        ("気温", "平均 (℃)", "平均 (℃)"),
    ]
)
# 列名の変更
df_HACHINOHE.columns = [
    "月",
    "降水量(mm)",
    "平均気温 (℃)",
]
# 行の抽出
df_HACHINOHE = df_HACHINOHE[df_HACHINOHE["月"].str.contains("月")]
# 型変換
df_HACHINOHE["降水量(mm)"] = df_HACHINOHE["降水量(mm)"].astype(float)
df_HACHINOHE["平均気温 (℃)"] = df_HACHINOHE["平均気温 (℃)"].astype(float)

df_HACHINOHE
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

# 最大値を取得して、Y軸の範囲を統一
max_value = max(
    df_HIROSAKI["降水量(mm)"].max(), df_HACHINOHE["降水量(mm)"].max()
)
# 最大値に少し余裕を持たせる
y_max = max_value * 1.1

figure, axes = plt.subplots(2, 1, figsize=(6, 6))
# 弘前市
axes[0].bar(df_HIROSAKI["月"], df_HIROSAKI["降水量(mm)"])
axes[0].set_title("弘前市")
axes[0].set_ylabel("降水量(mm)", fontsize=12)
axes[0].set_ylim(0, y_max)  # Y軸の範囲を統一

# 八戸市
axes[1].bar(df_HACHINOHE["月"], df_HACHINOHE["降水量(mm)"])
axes[1].set_title("八戸市")
axes[1].set_ylabel("降水量(mm)", fontsize=12)
axes[1].set_ylim(0, y_max)  # Y軸の範囲を統一

# レイアウト調整
plt.suptitle("弘前市と八戸市の月別降水量", fontsize=18)
plt.tight_layout()
plt.show()
```

```python
figure, axes = plt.subplots(figsize=(6, 4))
# 線グラフ
axes.plot(
    df_HIROSAKI["月"], df_HIROSAKI["平均気温 (℃)"], marker="o", label="弘前"
)
axes.plot(
    df_HACHINOHE["月"],
    df_HACHINOHE["平均気温 (℃)"],
    marker="x",
    label="八戸",
)
axes.set_ylabel("平均気温 (℃)", fontsize=12)
axes.set_title("弘前市と八戸市の月別平均気温", fontsize=18)

# レイアウト調整
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```
