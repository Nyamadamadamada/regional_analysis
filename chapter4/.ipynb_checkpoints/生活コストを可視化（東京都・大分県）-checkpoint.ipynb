## 4.5 生活コストを可視化（東京都・大分県）
東京都・大分県の生活コストデータは書籍を参考に取得してください。

```python
import pandas as pd

df_life_origin = pd.read_csv("./生活コスト.csv", encoding="shift-jis")
df_life_origin.head()
```

```python
df_life = df_life_origin.filter(
    items=[
        "収支項目分類（中分類）（購入形態用）94区分",
        "地域区分（都道府県）",
        "value",
    ]
)
df_life = df_life.rename(
    columns={
        "収支項目分類（中分類）（購入形態用）94区分": "品目分類",
        "地域区分（都道府県）": "地名",
        "value": "支出",
    }
)
df_life
```

```python
df_cross = df_life.pivot_table(
    index="品目分類", columns="地名", values="支出", sort=False
)
df_cross
```

```python
# 交通・通信から「交通」行を算出
df_cross.loc["交通"] = df_cross.loc["交通・通信"] - df_cross.loc["通信"]

# 不要な行を削除
df_cross = df_cross.drop(index=["交通・通信", "通信"])
df_cross
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import numpy as np

x_label = df_cross.columns  # X軸のラベル
x = np.arange(len(x_label))  # X軸の位置

figure, axes = plt.subplots(figsize=(6, 5))

axes.bar(
    x,
    df_cross.loc["消費支出"],
    width=0.4,
    facecolor=["tab:blue", "tab:orange"],  # 棒グラフの色
    hatch=["", "//"],  # 棒グラフのハッチ
)

# 軸・タイトル・凡例
axes.set_xticks(x)
axes.set_xticklabels(x_label, fontsize=16)
axes.set_ylabel("月支出（円）")
axes.set_title("一人暮らしの月支出の違い", fontsize=18)

plt.tight_layout()
plt.show()
```

```python
df_others = df_cross.drop(index="消費支出")

x_label = df_others.index  # X軸のラベル
x = np.arange(len(x_label))  # X軸の位置
width = 0.4  # 棒グラフの幅

figure, axes = plt.subplots(figsize=(8, 5))
axes.bar(x - width / 2, df_others["東京都"], width, label="東京都")
axes.bar(
    x + width / 2, df_others["大分県"], width, label="大分県", hatch="//"
)

# 軸・タイトル・凡例
axes.set_xticks(x)
axes.set_xticklabels(x_label, fontsize=16)
axes.set_ylabel("月支出（円）")
axes.set_title("一人暮らしの月支出の違い（品目別）", fontsize=18)
axes.legend(fontsize=16)

plt.tight_layout()
plt.show()
```

```python
# 差額を計算
df_cross["差額"] = df_cross["東京都"] - df_cross["大分県"]

# フォーマットを整えて表示
df_cross.style.format("{:,.0f}円")
```
