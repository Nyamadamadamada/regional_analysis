## 用途地域を可視化（函館市）

```python
import folium
from shapely.geometry import Point
import geopandas as gpd

data = [
    ("地点１（函館駅近く）", Point(140.730, 41.775)),
    ("地点２（五稜郭近く）", Point(140.760, 41.800)),
    ("地点３（香雪園近く）", Point(140.800, 41.790)),
]
target_points = gpd.GeoDataFrame(
    data, columns=["場所", "geometry"], crs=6668
)

map = folium.Map(
    location=[41.8, 140.75], zoom_start=12, tiles="Stadia.AlidadeSmooth"
)
# 地点をマーカーで表示
for _, row in target_points.iterrows():
    folium.Marker(
        location=[row.geometry.y, row.geometry.x],
        tooltip=row["場所"],
    ).add_to(map)

map
```

```python
zoning_area_origin = gpd.read_file("./函館_用途地域.geojson")
zoning_area = zoning_area_origin.filter(items=["用途地域", "geometry"])
zoning_area.head()
```

```python
print(zoning_area.crs)
```

```python
# 用途地域の分類マッピング
ZONING_CATEGORIES = {
    "第１種低層住居専用地域": "低層住居専用地域",
    "第２種低層住居専用地域": "低層住居専用地域",
    "第１種中高層住居専用地域": "中高層住居専用地域",
    "第２種中高層住居専用地域": "中高層住居専用地域",
    "第１種住居地域": "住居地域",
    "第２種住居地域": "住居地域",
    "準住居地域": "準住居地域",
    "近隣商業地域": "商業地域",
    "商業地域": "商業地域",
    "準工業地域": "工業地域",
    "工業地域": "工業地域",
    "工業専用地域": "工業地域",
}

# 用途地域の色マッピング
ZONING_COLORS = {
    "低層住居専用地域": "green",
    "中高層住居専用地域": "lightgreen",
    "住居地域": "yellow",
    "準住居地域": "blue",
    "商業地域": "pink",
    "工業地域": "gray",
}


def classify_zoning(zoning_type):
    """用途地域の大分類を取得する"""
    return ZONING_CATEGORIES.get(zoning_type)


def get_color(zoning_type):
    """用途地域の色を取得する"""
    return ZONING_COLORS.get(zoning_type, "white")


# 新しい列を追加
zoning_area["用途地域_大分類"] = zoning_area["用途地域"].map(classify_zoning)
zoning_area["color"] = zoning_area["用途地域_大分類"].map(get_color)

zoning_area.head()
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

zoning_area.plot(color=zoning_area["color"])
```

```python
def style_function(feature):
    """GeoJSONデータの各エリアのスタイルを設定"""
    color = feature["properties"].get("color")
    return {
        "fillColor": color,  # 塗りつぶしの色
        "fillOpacity": 0.6,  # 塗りの透明度
        "color": "black",  # 境界線の色
        "weight": 0.2,  # 境界線の太さ
    }

map = folium.Map(
    location=[41.8, 140.75], zoom_start=13, tiles="Stadia.AlidadeSmooth"
)

# ポイントのマーカー
for _, row in target_points.iterrows():
    folium.Marker(
        location=[row.geometry.y, row.geometry.x], tooltip=row["場所"]
    ).add_to(map)

# 用途地域レイヤー
folium.GeoJson(
    zoning_area,
    style_function=style_function,
    tooltip=folium.GeoJsonTooltip(fields=["用途地域_大分類"]),
).add_to(map)

map
```

```python
result = gpd.sjoin(target_points, zoning_area, predicate="within")
result[["場所", "用途地域"]]
```

## 最寄りのスーパーまでの距離を可視化（函館市）

```python
import osmnx as ox

supermarkets_origin = ox.features_from_place(
    "Hakodate, Hokkaido, Japan", tags={"shop": "supermarket"}
)
supermarkets_origin.head()
```

```python
supermarkets_origin["geometry"].geom_type.unique()
```

```python
supermarkets = supermarkets_origin.reset_index().filter(
    items=["name", "geometry"]
)
supermarkets = supermarkets.to_crs(epsg=3857)
supermarkets["geometry"] = supermarkets["geometry"].centroid
supermarkets.head()
```

```python
supermarkets.plot()
```

```python
target_points = target_points.to_crs(epsg=3857)

for _, row in target_points.iterrows():
    min_distance = supermarkets.geometry.distance(row.geometry).min()
    print(f"{row['場所']}の最も近いスーパーまでの距離: {min_distance:.0f}m")
```

```python
supermarkets["buffer_500m"] = supermarkets.geometry.buffer(500)
supermarkets.head()
```

```python
zoning_area = zoning_area.to_crs(epsg=3857)
print(zoning_area.crs)
```

```python
from matplotlib.patches import Patch

figure, axes = plt.subplots(figsize=(10, 8))
zoning_area.plot(ax=axes, edgecolor='lightgray', facecolor='lightgray')

# スーパーの位置
supermarkets.plot(
    ax=axes, color="blue", markersize=10, label="スーパーの位置"
)
# スーパーのバッファ
supermarkets.set_geometry("buffer_500m").plot(
    ax=axes, facecolor="none", edgecolor="blue", hatch="////", alpha=0.5
)
# 地点の位置
target_points.plot(
    ax=axes, color="red", marker="*",markersize=200, label="地点の位置"
)

# バッファの凡例を合体させる
handles, labels = axes.get_legend_handles_labels()
handles.insert(1, Patch(facecolor='none', edgecolor='blue', hatch='////'))
labels.insert(1, 'スーパーから半径500m')

plt.legend(handles, labels, fontsize=12)
plt.show()
```

```python
covered_rate = []

# 用途地域ごとに処理
for label, group in zoning_area.groupby("用途地域_大分類"):
    # 用途地域とバッファの交差面積を求める
    covered_area = gpd.overlay(
        group, supermarkets.set_geometry("buffer_500m")
    )
    total_area = group.area.sum()
    covered_area_sum = covered_area.area.sum()

    # カバー率の計算
    coverage_ratio = covered_area_sum / total_area * 100
    covered_rate.append((label, total_area / 10**6, coverage_ratio))

# DataFrameへ格納
df_covered = pd.DataFrame(
    covered_rate,
    columns=["用途地域_大分類", "総面積 (km²)", "スーパーのカバー率 (%)"],
)
df_covered
```
